```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
phool = sns.load_dataset('iris')
phool
```

```python
sns.lineplot(data=phool, x='sepal_length', y='sepal_width')
plt.title('Pholon ka plot')
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# Set styles
- darkgrid
- whitegrid
- dark
- white
- ticks


```python
sns.lineplot(data=phool, x='sepal_length', y='sepal_width')
plt.title('Pholon ka plot')
sns.set_style('dark')
plt.show()
```

# Size of the figure

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=phool, x='sepal_length', y='sepal_width')
plt.title('Pholon ka plot')
sns.set_style('dark')
plt.show()
```
