```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
phool = sns.load_dataset('iris')
phool
```

```python
plt.figure(figsize=(12,6))

sns.barplot(data=phool, x='sepal_length', y='sepal_width')

plt.title('Pholon ka plot')

plt.show()
```

```python
#plt.figure(figsize=(12,6))
sns.barplot(data=phool, x='species', y='sepal_width')
plt.title('Pholon ka plot')
sns.set_style('dark')
plt.show()
```

```python
sns.barplot(data=phool, x='species', y='petal_length')
plt.title('Pholon ka plot')
sns.set_style('dark')
plt.show()
```
